```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/mtcars.csv')
```

```python
df
```

```python
# Separate input and output variables
x = df[['disp','hp','wt']]

y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x, y)
```

```python
reg.coef_, reg.intercept_
```

```python
reg.score(x, y)
```

```python
disp = 221; hp=102; wt=3.91
```

```python
reg.predict([[221,125,2.91]])
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('disp vs. mpg')
plt.scatter(x['disp'], y)
plt.subplot(2,2,2)
plt.title('hp vs. mpg')
plt.scatter(x['hp'], y)
plt.subplot(2,2,3)
plt.title('wt vs. mpg')
plt.scatter(x['wt'], y)
```
