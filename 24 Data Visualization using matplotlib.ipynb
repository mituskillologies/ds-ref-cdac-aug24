##### Bar plot

```python
student = pd.read_csv('datasets/student.csv')
```

```python
plt.figure(figsize=(16,9))
plt.bar(student['name'], student['marks'], align='edge',color = 'navy',
       label='marks')
plt.bar(student['name'], student['age'], color = 'red', label = 'age')
plt.legend()
```

```python
plt.figure(figsize=(16,9))
plt.barh(student['name'], student['marks'], align='edge',color = 'aqua',
       label='marks')
plt.barh(student['name'], student['age'], color = 'pink', label = 'age')
plt.legend()
```
