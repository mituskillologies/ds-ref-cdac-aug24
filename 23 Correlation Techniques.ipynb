```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([3,15,17,34,38,45,59,61,78,81])
y = np.array([10,14,23,33,31,38,45,58,66,71])
z = np.array([89,86,77,72,56,68,53,42,32,29])
p = np.array([67,56,44,57,76,45,67,44,25,67])
```

```python
np.corrcoef(x, y)
```

```python
np.corrcoef(x, z)
```

```python
np.corrcoef(x, p)
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(x, y)
```

```python
pearsonr(x, z)
```

```python
pearsonr(x, p)
```
